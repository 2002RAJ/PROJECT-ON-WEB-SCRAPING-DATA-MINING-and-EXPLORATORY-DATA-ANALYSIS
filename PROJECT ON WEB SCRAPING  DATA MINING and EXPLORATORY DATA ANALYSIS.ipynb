```python
soup=BeautifulSoup(page.text,"html.parser")
box = soup.find("div",class_="_1YokD2 _3Mn1Gg")
Brand = []
for i in box.find_all("div",class_="_4rR01T"):   
    text = i.text
    regex = re.findall('(\w+\s\w+\s\w+\s\w+|\w+\s\w+\s\w+|\w+\s\w+|\w+)',text)
    if regex:
        Brand.append(regex[0])
    else:
        Brand.append(np.nan)
print(Brand) 
print(len(Brand))
```

```python
Brand = []
Price = []
color = []
RAM = []
storage = []
display = []
camera = []
battery = []
Ratings = []

for i in range(1,80):
    url=f"https://www.flipkart.com/search?q=mobile&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off&page={i}"
    page=requests.get(url)
    soup=BeautifulSoup(page.text)
    
    soup=BeautifulSoup(page.text,"html.parser")
    
    for i in soup.find_all("div",class_="_4rR01T"):   
        text = i.text
        regex = re.findall('(\w+\s\w+\s\w+\s\w+|\w+\s\w+\s\w+|\w+\s\w+|\w+)',text)
        if regex:
            Brand.append(regex[0])
        else:
            Brand.append(np.nan)
            
    for i in soup.find_all("div",class_="_4rR01T"):   
        text = i.text
        regex = re.findall('(\w+|\w+\s\w+),',text)
        if regex:
            color.append(regex[0])
        else:
            color.append(np.nan)
            
    #this code is for ratings
    for i in soup.find_all("div",class_="_3LWZlK"):   
        Ratings.append(i.text)
     
    #this code is for price
    for i in soup.find_all("div",class_="_30jeq3 _1_WHN1"):   
        Price.append(i.text)
     
    #this code is for ROM
    for i in soup.find_all("div",class_="_4rR01T"):   
        text = i.text
        regex = re.findall('\d+\sGB',text)
        if regex:
            storage.append(regex[0])
        else:
            storage.append(np.nan)
     
    #this code is for camera
    for i in soup.find_all("ul",class_="_1xgFaf"):   
        text = i.text
        regex = re.findall('Display(\w+\s\+\s\w+\s\+\s\w+|\w+\s\+\s\d+\.\w+|\w+\s\+\s\w+|\w+)',text)
        if regex:
            camera.append(regex[0])
        else:
            camera.append(np.nan)
     
    #this code is for RAM
    for i in soup.find_all("ul",class_="_1xgFaf"):   
        text = i.text
        regex = re.findall('(\d+\sGB)\sRAM',text)
        if regex:
            RAM.append(regex[0])
        else:
            RAM.append(np.nan)
    
    #this code is for display
    for i in soup.find_all("ul",class_="_1xgFaf"):   
        text = i.text
        regex = re.findall('(\d+.\d+)\sinch',text)
        if regex:
            display.append(regex[0])
        else:
            display.append(np.nan)
    
    #this code is for battery
    for i in soup.find_all("ul",class_="_1xgFaf"):   
        text = i.text
        regex = re.findall('(\d+)\smAh',text)
        if regex:
            battery.append(regex[0])
        else:
            battery.append(np.nan)
```
